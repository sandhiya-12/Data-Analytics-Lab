```
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
```

```
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.model_selection import train_test_split
```

```
iris = load_iris()
```

```
print(iris.feature_names)#attributes
print(iris.target_names)#class labels
print(iris.DESCR)#data set characteristics
print(iris.data)#prints data of iris
```

```
input_data = iris.data #give iris dataset as input
```

```
target_class = iris.target #assign integer values to class labels
iris['target']
print(iris.target)
```

```
scaler = StandardScaler() #z-score normalizer
input_scaled = scaler.fit_transform(input_data) #apply z-score norm to input data
print(input_scaled)
```

```
# One hot encoding
enc = OneHotEncoder() #convert categorical data to numeric array
hot_target_class = enc.fit_transform(target_class[:, np.newaxis]).toarray() #converts target names into numeric array
print(hot_target_class)
```

```
x_train,x_test,y_train,y_test = train_test_split(input_scaled,hot_target_class,test_size=0.3) #split the input and target data into training and testing
x_train
x_test
```

```
model = Sequential() #creating sequential model
```

```
#add layers to the model using dense func
h1 = model.add(Dense(6,activation="relu",input_dim=4))
h2=model.add(Dense(5,activation="relu"))
h3=model.add(Dense(7,activation="relu"))
o = model.add(Dense(3,activation="softmax"))
```

```
#configuring the learning process
model.compile(loss="categorical_crossentropy",optimizer="Adam",metrics=['accuracy'])
```

```
#print the summary of the model
model.summary()
```

```
#fit the training data to the generated model
model.fit(x_train,y_train,batch_size=10,epochs=50)
```

```
 #predict the test data 
  score = model.evaluate(x_test,y_test, verbose=0)
 print('Test loss:', score[0])
 print('Test accuracy:', score[1])
```
