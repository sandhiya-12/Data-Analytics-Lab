```python
dat=pd.read_csv("C:\\Users\\Sandy\\Downloads\\train.csv")
df=pd.DataFrame(dat)
```

```python
print(df)
```

```python
df.info()
```

```python
df['Gender']=df['Gender'].fillna("Any")
df.info()
```

```python
df['Dependents']=df['Dependents'].fillna("0")
df.info()
```

```python
df['Married']=df['Married'].fillna("No")
df.info()
```

```python
df['Self_Employed']=df['Self_Employed'].fillna("No info")
df.info()
```

```python
df['LoanAmount']=df['LoanAmount'].fillna(df['LoanAmount'].mean())
df.info()
```

```python
df['Loan_Amount_Term']=df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mean())
df.info()
```
